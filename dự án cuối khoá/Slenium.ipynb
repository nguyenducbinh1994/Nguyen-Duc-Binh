```python
import pandas as pd
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from tkinter import Tk, Label, Entry, Button, filedialog, messagebox

def read_emails_from_excel(file_path):
    """Đọc email và tên người nhận từ tệp Excel."""
    try:
        df = pd.read_excel(file_path)
        emails = df['E-MAIL'].dropna().tolist()  
        importers = df['IMPORTER NAME'].dropna().tolist()
        return emails, importers
    except Exception as e:
        messagebox.showerror("Error", f"Error reading Excel file: {e}")
        return [], []

def send_email(subject, body, sender, recipient, password):
    """Gửi email với nội dung HTML."""
    try:
        msg = MIMEMultipart()
        msg['Subject'] = subject
        msg['From'] = sender
        msg['To'] = recipient

        msg.attach(MIMEText(body, 'html'))

        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
            smtp_server.login(sender, password)
            smtp_server.sendmail(sender, recipient, msg.as_string())

        print(f"Message sent to {recipient}!")
    except Exception as e:
        print(f"Failed to send email to {recipient}: {e}")

def select_file(entry_widget, filetypes):
    """Mở hộp thoại để chọn tệp và hiển thị đường dẫn trong Entry widget."""
    file_path = filedialog.askopenfilename(filetypes=filetypes)
    entry_widget.delete(0, "end")
    entry_widget.insert(0, file_path)

def load_template(file_path):
    """Đọc nội dung tệp HTML mẫu."""
    try:
        with open(file_path, encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        messagebox.showerror("Error", "Error: 'body.html' file not found.")
    except Exception as e:
        messagebox.showerror("Error", f"Error reading HTML template: {e}")
    return ""

def send_bulk_emails():
    """Gửi email hàng loạt đến tất cả các địa chỉ trong tệp Excel."""
    excel_file = excel_entry.get()
    sender_email = sender_entry.get()
    email_password = password_entry.get()
    html_file = html_entry.get()

    if not excel_file or not sender_email or not email_password:
        messagebox.showwarning("Input Error", "Please provide all the required inputs!")
        return

    template = load_template(html_file)
    emails, importers = read_emails_from_excel(excel_file)

    if not emails or not importers:
        messagebox.showerror("Error", "No emails or importers found.")
        return

    for email, importer in zip(emails, importers):
        body = template.format(ten_cong_ty=importer) if template else f"Dear {importer},\n\nThis is a test email."
        send_email(subject_entry.get(), body, sender_email, email, email_password)

    messagebox.showinfo("Success", "Emails sent successfully!")

# Tạo giao diện người dùng bằng tkinter
root = Tk()
root.title("Email Sender")

def create_label_entry(row, label_text, entry_width=50, show_text=None):
    """Hàm tạo các Label và Entry với cấu hình chung."""
    Label(root, text=label_text).grid(row=row, column=0, padx=10, pady=5, sticky="e")
    entry = Entry(root, width=entry_width, show=show_text)
    entry.grid(row=row, column=1, pady=5)
    return entry

# Labels và Entries cho giao diện
excel_entry = create_label_entry(0, "Excel File:")
Button(root, text="Browse", command=lambda: select_file(excel_entry, [("Excel files", "*.xlsx;*.xls")])).grid(row=0, column=2, padx=10, pady=5)

sender_entry = create_label_entry(1, "Sender Email:")
password_entry = create_label_entry(2, "Email Password:", show_text="*")

html_entry = create_label_entry(3, "HTML Template File:")
Button(root, text="Browse", command=lambda: select_file(html_entry, [("HTML files", "*.html")])).grid(row=3, column=2, padx=10, pady=5)

subject_entry = create_label_entry(4, "Email Subject:")

Button(root, text="Send Emails", command=send_bulk_emails).grid(row=5, column=0, columnspan=3, pady=20)

root.mainloop()
```
